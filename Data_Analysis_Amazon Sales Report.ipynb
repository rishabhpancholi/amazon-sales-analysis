# **Data Cleaning**

**1.Treating wrong data**

```python
for col in ['Status', 'Fulfilment', 'Sales Channel', 'Ship Service Level', 'Category', 'Size', 'Courier Status', 'Qty',
            'Currency', 'Amount', 'Ship City', 'Ship State', 'Ship Postal Code', 'Ship Country', 
            'B2B', 'Fullfilled By', 'New', 'PendingS']:
  print(col, ':')
  print(df[col].unique().tolist())
  print('\n')
```

Ship City --> has values in wrong format and have wrong data too

Ship State --> has values in wrong format and have wrong data too

```python
df['Ship City'] = df['Ship City'].str.title()
```

```python
df['Ship State'] = df['Ship State'].str.title()
```

```python
dic = {
  'Chandigarh' : 'Punjab',
  'Rajshthan' : 'Rajasthan',
  'Nl' : 'Nagaland',
  'New Delhi' : 'Delhi',
  'Punjab/Mohali/Zirakpur' : 'Punjab',
  'Rj' : 'Rajasthan',
  'Orissa' : 'Odisha',
  'Pb' : 'Punjab',
  'Apo' : 'Andhra Pradesh',
  'Ar' : 'Arunachal Pradesh',
  'Pondicherry' : 'Puducherry',
  'Rajsthan' : 'Rajasthan'
}

df['Ship State'] = df['Ship State'].replace(dic)
```

**2. Treating missing values**

```python
df.isnull().sum() / len(df) * 100
```

```python
df = df.drop(columns = {'Fullfilled By', 'New', 'PendingS'})
```

```python
df['Currency'] = df['Currency'].fillna(df['Currency'].mode()[0])
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].median())
```

```python
df[df.isnull().any(axis = 1)]
```

```python
df = df.dropna(subset = ['Ship City'])
```

```python
df.isnull().sum()
```

**3. Treating wrong datatypes**

```python
df['Ship Postal Code'] = df['Ship Postal Code'].astype(int)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.sort_values(by = ['Date'], ignore_index = True)
```

```python
df.dtypes
```

```python
df
```

**4. Treating Duplicates**

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

**5. Treating Outliers**

Retaining Outliers

**6. Dimension Reduction**

Dropping "Ship Country" and "Currency" as they have only one unique value

```python
df = df.drop(columns = {'index', 'Order ID', 'Currency', 'Ship Country'})
df
```

**Data Cleaning is completed**

```python
df.to_csv(r'D:\Study\Data Science\Data Analytics\Data Analytics Practice Project\Cleaned Datasets\Cleaned_Amazon Sales Report.csv', index = False)
```

**Q1. What are the unique categories in 'Status', 'Fulfilment', 'Sales Channel', 'Ship Service Level', 'Qty', 'Category', 'Size', 'Courier Status', 'Ship State', 'B2B' and their counts?**

```python
for col in ['Status', 'Fulfilment', 'Sales Channel', 'Ship Service Level', 'Category', 'Size', 'Courier Status', 'Qty',
            'Ship State', 'B2B']:
  if(col == 'Ship State'):
    plt.figure(figsize = (16, 10))
    sns.countplot(x = df[col])
    plt.xticks(rotation = 90)
    plt.show()
  else:
    sns.countplot(x = df[col])
    plt.xticks(rotation = 90)
    plt.show()
```

**Q2. How is "Status" related to other discrete variables?**

```python
for col in ['Fulfilment', 'Sales Channel', 'Ship Service Level', 'Category', 'Size', 'Courier Status', 'Qty', 'Ship State',
            'B2B']:
  plt.figure(figsize = (16, 10))
  sns.countplot(x = df['Status'], hue = df[col])
  plt.xticks(rotation = 90)
  plt.show()
```
